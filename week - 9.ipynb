```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
```

```python
iris = datasets.load_iris()

# Create DataFrame for features
X = pd.DataFrame(iris.data, columns=['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width'])

# Create DataFrame for target labels
y = pd.DataFrame(iris.target, columns=['Target'])
```

```python
# Define color map for 3 classes
colormap = np.array(['red', 'lime', 'black'])

plt.figure(figsize=(14, 7))

# Sepal plot
plt.subplot(1, 2, 1)
plt.scatter(X.Sepal_Length, X.Sepal_Width, c=colormap[y.Target], s=40)
plt.title('Sepal - True Classification')

# Petal plot
plt.subplot(1, 2, 2)
plt.scatter(X.Petal_Length, X.Petal_Width, c=colormap[y.Target], s=40)
plt.title('Petal - True Classification')
plt.show()
```

```python
# Initialize and fit KMeans model
model = KMeans(n_clusters=3, random_state=42)
model.fit(X)

# Print predicted cluster labels
print("KMeans Cluster Labels:\n", model.labels_)
```

```python
plt.figure(figsize=(14, 7))

# Real classification (Petal)
plt.subplot(1, 2, 1)
plt.scatter(X.Petal_Length, X.Petal_Width, c=colormap[y.Target], s=40)
plt.title('Petal - Real Classification')

# KMeans classification
plt.subplot(1, 2, 2)
plt.scatter(X.Petal_Length, X.Petal_Width, c=colormap[model.labels_], s=40)
plt.title('Petal - KMeans Classification')
plt.show()
```
